Model XLM-Roberta

```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# 1. CSV faylını oxu
df2 = pd.read_csv("pairs.csv")

# 2. Model və tokenizer yüklə
xlm_name = "cardiffnlp/twitter-xlm-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(xlm_name)
model = AutoModelForSequenceClassification.from_pretrained(xlm_name)

# 3. Sentiment mapping funksiyası
def map_sentiment(pred_id):
    if pred_id == 0:
        return "negative"
    elif pred_id == 1:
        return "neutral"
    else:
        return "positive"

# 4. Cümlələri al
texts = df2["new"].tolist()

# 5. Tokenize et
inputs = tokenizer(texts, padding=True, truncation=True, return_tensors="pt")

# 6. Proqnoz (inference)
with torch.no_grad():
    outputs = model(**inputs)
    preds = torch.argmax(outputs.logits, dim=1)  # burada preds yaradılır

# 7. Nəticəni df2-yə əlavə et
df2["xlm_pred_id"] = preds.numpy()
df2["xlm_pred"] = df2["xlm_pred_id"].apply(map_sentiment)

# 8. Yoxla
print(df2[["new", "xlm_pred"]].head())
```

Model Multilingual BERT

```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# CSV faylı oxu
df2 = pd.read_csv("pairs.csv")  # səninki CSV adı ilə əvəz et

# Model və tokenizer
bert_name = "nlptown/bert-base-multilingual-uncased-sentiment"
bert_tokenizer = AutoTokenizer.from_pretrained(bert_name)
bert_model = AutoModelForSequenceClassification.from_pretrained(bert_name)

# Sentiment mapping (bu modeldə 1-5 arası skor verir)
def map_bert_sentiment(score):
    if score <= 2:
        return "negative"
    elif score == 3:
        return "neutral"
    else:
        return "positive"

# Cümlələri al
texts = df2["new"].tolist()
inputs = bert_tokenizer(texts, padding=True, truncation=True, return_tensors="pt")

# Proqnoz
with torch.no_grad():
    outputs = bert_model(**inputs)
    logits = outputs.logits
    preds = torch.argmax(logits, dim=1)  # ən yüksək ehtimallı class
    # model 0-4 label verir, 1-5 aralığında saymaq üçün 1 əlavə edə bilərik
    scores = preds + 1

# Nəticəni df2-yə əlavə et
df2["bert_pred_id"] = scores.numpy()
df2["bert_pred"] = df2["bert_pred_id"].apply(map_bert_sentiment)

# Nəticəni gör
print(df2[["new", "bert_pred"]].head())

# CSV-yə saxlamaq
df2.to_csv("pairs_with_bert_sentiment.csv", index=False, encoding="utf-8")
```

```python
import pandas as pd
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from sklearn.metrics import classification_report

# CSV faylını oxu
df2 = pd.read_csv("pairs.csv")  # Sənin CSV faylının adı
df2.head()

# Sentiment mapping
label2id = {"negative": 0, "neutral": 1, "positive": 2}
id2label = {v: k for k, v in label2id.items()}

# sentiment_id sütununu yarat
df2["sentiment_id"] = df2["sentiment"].map(label2id)

# XLM-RoBERTa
xlm_name = "cardiffnlp/twitter-xlm-roberta-base-sentiment"
xlm_tokenizer = AutoTokenizer.from_pretrained(xlm_name)
xlm_model = AutoModelForSequenceClassification.from_pretrained(xlm_name)

texts = df2["new"].tolist()
inputs = xlm_tokenizer(texts, padding=True, truncation=True, return_tensors="pt")

with torch.no_grad():
    outputs = xlm_model(**inputs)
    probs = torch.softmax(outputs.logits, dim=1)
    preds = torch.argmax(probs, dim=1)

df2["xlm_pred_id"] = preds.numpy()
df2["xlm_pred"] = df2["xlm_pred_id"].map(id2label)

# Multilingual BERT
bert_name = "nlptown/bert-base-multilingual-uncased-sentiment"
bert_tokenizer = AutoTokenizer.from_pretrained(bert_name)
bert_model = AutoModelForSequenceClassification.from_pretrained(bert_name)

inputs_bert = bert_tokenizer(texts, padding=True, truncation=True, return_tensors="pt")

# Modeldən çıxış al
with torch.no_grad():
    outputs_bert = bert_model(**inputs_bert)
    probs_bert = torch.softmax(outputs_bert.logits, dim=1)
    preds_bert = torch.argmax(probs_bert, dim=1)

# Add BERT predictions to df2
df2["bert_pred_id"] = preds_bert.numpy()
df2["bert_pred"] = df2["bert_pred_id"].map(id2label)

# 5 sinifi 3 sinifə map et
def map_bert_to_3class(x):
    if x <= 1:       # 0,1 → negative
        return 0
    elif x == 2:     # 2 → neutral
        return 1
    else:            # 3,4 → positive
        return 2

df2["bert_pred_id_3class"] = df2["bert_pred_id"].apply(map_bert_to_3class)
# 3 sinifə map etdikdən sonra BERT pred üçün doğru map
df2["bert_pred"] = df2["bert_pred_id_3class"].map(id2label)

# 3 sinif target_names
target_names_3class = ["negative", "neutral", "positive"]

# Qiymətləndirmə
print("Multilingual BERT results (3 class):")
print(classification_report(df2["sentiment_id"], df2["bert_pred_id_3class"], target_names=target_names_3class))

# Nəticələri yoxlamaq üçün
print(df2[["new", "sentiment", "xlm_pred", "bert_pred"]].head())
```
